# Question 19 : Écrivez les requêtes SQL permettant de connaître…
# Les 10 pays ayant le plus haut ratio disponibilité alimentaire/habitant en termes de protéines (en kg)
# par habitant, puis en termes de kcal par habitant.

a) SELECT sum(dispo_alim.dispo_alim_kcal_p_j)*365, pays

FROM dispo_alim
GROUP BY pays

ORDER BY dispo_alim.dispo_alim_kcal_p_j DESC

LIMIT 10;

SELECT (sum(dispo_prot)* 0.001 * 365) AS dispo_prot_kg, pays

FROM dispo_alim
GROUP BY pays

ORDER BY dispo_prot_kg DESC

LIMIT 10;


# Pour l'année 2013, les 10 pays ayant le plus faible ratio disponibilité alimentaire/habitant
# en termes de protéines (en kg) par habitant.

b)SELECT pays, sum(dispo_alim.dispo_alim_kcal_p_j)*365

FROM dispo_alim

GROUP BY pays

ORDER BY dispo_alim.dispo_alim_kcal_p_j ASC

LIMIT 10;


SELECT pays, sum(dispo_alim.dispo_prot*0.001)*365 as somme_proteine

FROM dispo_alim 

GROUP BY pays

ORDER BY somme_proteine ASC

ASC LIMIT 10;


# La quantité totale (en kg) de produits perdus par pays en 2013.

SELECT pays, SUM(equilibre_prod.pertes)*1000000 as pertes

FROM equilibre_prod

GROUP BY pays;


# Les 10 pays pour lesquels la proportion de personnes sous-alimentées est la plus forte.

d) SELECT sous_nutrition.pays, sous_nutrition.nb_personnes, population.population, 
(sous_nutrition.nb_personnes / population.population) as ratio

FROM sous_nutrition 

LEFT JOIN population 

ON sous_nutrition.code_pays = population.code_pays


ORDER BY ratio DESC                

LIMIT 10;


# Les 10 produits pour lesquels le ratio Autres utilisations/Disponibilité intérieure 
# est le plus élevé.

e) SELECT produit, sum(dispo_int), sum(autres_utilisations), 
(sum(autres_utilisations)/sum(dispo_int)) as ratio

FROM equilibre_prod

GROUP BY produit

ORDER BY ratio DESC

LIMIT 10;
