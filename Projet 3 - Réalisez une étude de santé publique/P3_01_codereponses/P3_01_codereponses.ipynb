# Question 3 : Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.

## Vous ferez cela à partir de ces informations
     Population de chaque pays ;
     Disponibilité alimentaire donnée pour chaque produit et pour 
     chaque pays en kcal/personne/jour.
     Disponibilité alimentaire en protéines donnée pour chaque      
     produit et pour chaque pays en g/personne/jour.



```python
# Importer le fichier fr_animaux.csv
animaux = pd.read_csv('fr_animaux.csv', sep=',')
animaux = animaux.assign(source = 'animale')
animaux = animaux[animaux.Zone != 'Chine']

# Importer le fichier fr_vegetaux.csv
vegetaux = pd.read_csv('fr_vegetaux.csv', sep=',')
vegetaux = vegetaux.assign(source = 'végétale')
vegetaux = vegetaux[vegetaux.Zone != 'Chine']

# Union de 2 dataframes animaux et vegetaux
animaux_vegetaux = pd.concat([animaux, vegetaux])
#animaux_vegetaux =animaux_vegetaux[['Code zone','Zone', 'Code Produit','Produit','Code année','Élément', 'Valeur', 'source']]
animaux_vegetaux
```

```python
# Essayons d'obtenir la nourriture de chaque pays et de chaque produit 
nourriture = animaux_vegetaux[animaux_vegetaux.Élément == 'Nourriture']
nourriture = nourriture.assign(nourriture = nourriture['Valeur']*1000*1000)
nourriture = nourriture[['Code zone','Zone', 'Produit', 'Élément', 'nourriture', 'source']]
nourriture
```

```python
# Création d'un dataframe avec 2 colonnes zone et population en millier
population_sans_chine = population_sans_chine[['Zone', 'population_en_millier']]

# Calculons la disponibilité alimentaire en quantité (kg/personne/an) de chaque pays et de chaque produit 
dispo_alim_kg = animaux_vegetaux[animaux_vegetaux.Élément == 'Disponibilité alimentaire en quantité (kg/personne/an)']
dispo_alim_kg = dispo_alim_kg.assign(dispo_alim_kg_personne = dispo_alim_kg['Valeur'])
dispo_alim_kg
```

```python

# joindre le dataframe dispo_alim_kg au dataframe population_sans_chine sur la clé Zone avec la jointure à gauche
dispo_alim_kg = dispo_alim_kg.merge(population_sans_chine, how ='left', on = 'Zone')
dispo_alim_kg
```

```python

# création de la colonne dispo_alim_kg d'un produit pour la population d'un pays donné 
dispo_alim_kg = dispo_alim_kg.assign(dispo_alim_kg = dispo_alim_kg['dispo_alim_kg_personne']*dispo_alim_kg['population_en_millier'])
dispo_alim_kg
```

```python
# creation du dataframe de disponibilite alimentaire kcal
dispo_alim_kcal = animaux_vegetaux[animaux_vegetaux.Élément == 'Disponibilité alimentaire (Kcal/personne/jour)']

# jointure de dispo_alim_kcal et population_sans_chine à gauche et sur la clé Zone
dispo_alim_kcal = dispo_alim_kcal.merge(population_sans_chine, how='left', on = 'Zone')

#création de la colonnne dispo_alim_kcal qui représente la dispo_alim_kcal de chaque produit par an par pays
dispo_alim_kcal = dispo_alim_kcal.assign(dispo_alim_kcal = dispo_alim_kcal['Valeur'] * dispo_alim_kcal['population_en_millier'] * 365)
dispo_alim_kcal
```

```python
# Affichage de la disponibilité de protéines en quantité (g/personne/jour)
# calcul de la disponibilité de protéines en quantité (g/personne/jour) de chaque pays et de chaque produit
dispo_prot_g = animaux_vegetaux[animaux_vegetaux.Élément == 'Disponibilité de protéines en quantité (g/personne/jour)']

# Jointure à gauche entre dispo_prot_g et population_sans_chine sur la clé Zone
dispo_prot_g = dispo_prot_g.merge(population_sans_chine, how = 'left', on ='Zone')
# convertir la disponibilité alimentaire en kg/personne/an
dispo_prot_kg = dispo_prot_g.assign(dispo_prot_kg = (dispo_prot_g['Valeur']* 0.001 * dispo_prot_g['population_en_millier']*365))
dispo_prot_kg
```

# Question 4 : A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire (pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg. Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio aux données disponibles sur internet, par exemple en cherchant la valeur calorique d'un oeuf.

```python
# Creation du ratio: jointure à gauche entre dispo_alim_kcal et dispo_alim_kg sur les 2 clés Zone et Produit
ratio = dispo_alim_kcal.merge(dispo_alim_kg, on=['Zone', 'Produit'], how='left')
# Jointure entre ratio et dispo_prot_kg à gauche sur les clés Zone et Produit
ratio = ratio.merge(dispo_prot_kg,on=['Zone', 'Produit'], how='left')
# Sélection des colonnes ci-dessous
ratio = ratio[['Code zone','Zone','Produit','Élément', 'dispo_alim_kcal', 'dispo_prot_kg', 'dispo_alim_kg', 'source']]
# Calcul du ratio energie_sur_poids
ratio = ratio.assign(energie_sur_poids = ratio['dispo_alim_kcal'] / ratio['dispo_alim_kg'])
# Remplaçons les cellules en NAN par 0
ratio['energie_sur_poids'].fillna(0, inplace=True)
# affichage des 5 premieres lignes
ratio.head(5)
```

```python
# comparer à une valeur de référence de l'oeuf sur wiki
# Sur wikipedia œuf moyen de 100 g : 147 kcal calories => ratio :1470 

# selectionner les oeufs parmi les produits
oeuf = ratio[ratio.Produit == 'Oeufs']
# Choix des oeufs de la france
oeuf = oeuf[oeuf.Zone == 'France']
# projection sur les colonnes ci-dessous
oeuf[['Zone', 'Produit', 'energie_sur_poids']]
```

# En suivant la même méthodologie, calculez également le pourcentage de protéines de chaque produit (pour chaque pays). Ce pourcentage est obtenu en calculant le ratio "poids de protéines/poids total" (attention aux unités utilisées)

```python
# Calcul du ratio poids de protéines sur poids sur poids total
ratio = ratio.assign(proteine_sur_poids = ratio['dispo_prot_kg'] / ratio['dispo_alim_kg'])
ratio['proteine_sur_poids'].fillna(0, inplace=True)
ratio
```

```python
# "Sécurité alimentaire" columns
sa_columns = ['Code zone', 'Zone', 'Élément', 'Produit', 'Année', 'Valeur']

sousalimentation = pd.read_csv("fr_sousalimentation.csv")[sa_columns]

sousalimentation.rename(
    columns={
        'Code zone': 'code_zone',
        'Zone': 'zone',
        'Valeur': 'pop_sous_alim',
        'Année': 'annee'
    },
    inplace=True
)

def year_format(year_code):
    """ From year_code format (XXXXXXX) to standard year format (XXXX) """
    
    if year_code == '2011-2013':
        return 2012

    elif year_code == '2012-2014':
        return 2013
    
    elif year_code == '2013-2015':
        return 2014
    
    elif year_code == '2014-2016':
        return 2015
    
    elif year_code == '2015-2017':
        return 2016
    
    elif year_code == '2016-2018':
        return 2017

# Apply the standard year format XXXX
sousalimentation['annee'] = sousalimentation['annee'].apply(year_format)

sousalimentation.fillna(0, inplace=True)
sousalimentation['pop_sous_alim'] = pd.to_numeric(sousalimentation['pop_sous_alim'], errors='coerce')

sousalimentation = sousalimentation.assign(pop_sous_alim =  sousalimentation['pop_sous_alim']* 1000000)


# Enlever "China" (We just keep the China sub-countries)
sousalimentation = sousalimentation[sousalimentation.zone != 'Chine'].copy()

# sousalimentaion en 2013
sousalimentation_2013 = sousalimentation[sousalimentation.annee == 2013]

sousalimentation_2013.head(5)
```

```python
# Rename
animaux_vegetaux.rename(columns={
    'Code zone': 'code_zone',
    'Zone': 'zone',
    'Code Produit': 'code_produit',
    'Produit': 'produit',
    'Année': 'annee',
    'source' : 'origin',
    'Aliments pour animaux': 'alim_anim',
    'Autres utilisations (non alimentaire)': 'autre_util',
    'Disponibilité alimentaire (Kcal/personne/jour)': 'dispo_alim_kcal',
    'Disponibilité alimentaire en quantité (kg/personne/an)': 'dispo_alim_quantite',
    'Disponibilité de matière grasse en quantité (g/personne/jour)': 'dispo_alim_grasse',
    'Disponibilité de protéines en quantité (g/personne/jour)': 'dispo_prot',
    'Exportations - Quantité': 'exportation',
    'Importations - Quantité': 'importation',
    'Nourriture': 'nour',
    'Pertes': 'perte',
    'Production': 'production',
    'Semences': 'semence',
    'Traitement': 'traitement',
    'Variation de stock': 'variation_stock',
    'Disponibilité intérieure': 'dispo_int'
}, inplace=True)

# Left join with the population size
animaux_vegetaux = animaux_vegetaux.merge(population_sans_chine, how='left', on='zone')

# Clean
animaux_vegetaux.fillna(0, inplace=True)

# From 1000 tons to kg
animaux_vegetaux['exportation'] = animaux_vegetaux['exportation'] * 1000000
animaux_vegetaux['autre_util'] = animaux_vegetaux['autre_util'] * 1000000
animaux_vegetaux['alim_anim'] = animaux_vegetaux['alim_anim'] * 1000000
animaux_vegetaux['importation'] = animaux_vegetaux['importation'] * 1000000
animaux_vegetaux['nour'] = animaux_vegetaux['nour'] * 1000000
animaux_vegetaux['perte'] = animaux_vegetaux['perte'] * 1000000
animaux_vegetaux['production'] = animaux_vegetaux['production'] * 1000000
animaux_vegetaux['semence'] = animaux_vegetaux['semence'] * 1000000
animaux_vegetaux['traitement'] = animaux_vegetaux['traitement'] * 1000000
animaux_vegetaux['variation_stock'] = animaux_vegetaux['variation_stock'] * 1000000
animaux_vegetaux['dispo_int'] = animaux_vegetaux['dispo_int'] * 1000000
animaux_vegetaux.head(10)
```

```python
animaux_vegetaux['dispo_alim_kcal_p_j'] = animaux_vegetaux['dispo_alim_kcal']
animaux_vegetaux['dispo_prot_p_j'] = animaux_vegetaux['dispo_prot']

animaux_vegetaux['dispo_alim_kcal'] = animaux_vegetaux['dispo_alim_kcal'] * animaux_vegetaux['population']
animaux_vegetaux['dispo_prot'] = animaux_vegetaux['dispo_prot'] * animaux_vegetaux['population']

animaux_vegetaux['dispo_alim_kcal'].fillna(0, inplace=True)
animaux_vegetaux['dispo_prot'].fillna(0, inplace=True)

display(animaux_vegetaux.head(10))
```

```python
# ajouter la colonne "est_sousalimente" au principal dataframe

temp_df = pd.DataFrame(sousalimentation_2013, columns=['code_zone', 'sousalimentation', 'est_sousalimente'])
animaux_vegetaux = animaux_vegetaux.merge(temp_df, how='left', on='code_zone')
animaux_vegetaux.head(50)
```

```python
# Dataframe with just undernourishment countries
animaux_vegetaux_11 = animaux_vegetaux[animaux_vegetaux['est_sousalimente'] == True].copy()

# Top 15 Export products
top_export_products = animaux_vegetaux.sort_values(by='exportation', ascending=False)[['code_produit', 'produit']].drop_duplicates()[0:15]
top_export_products_name = [x for x in top_export_products['produit']]
top_export_products_code = [x for x in top_export_products['code_produit']]

# Print them
print(""" Les 15 produits les plus exportés par les pays qui ont la plus grande population sousalimentée:

""")
for i, produit in enumerate(top_export_products_name):
    print(f"{i+1} - {produit}")
    
```

```python

# 200 produits les plus importés par les pays les plus sous-alimentés
top_import_of_poor_exports = animaux_vegetaux.loc[(animaux_vegetaux['code_produit'].isin(top_export_products_code)), :]

top_import_of_poor_exports = top_import_of_poor_exports.sort_values(by='importation', ascending=False)[0:200]

final_columns = ['code_produit', 'produit', 'alim_anim', 'autre_util', 'dispo_int', 'nour']
df_200 = top_import_of_poor_exports[final_columns].copy()
#df_q11 = animaux_vegetaux[final_columns].copy()
df_200 = df_200.groupby(['code_produit', 'produit']).sum()
df_200.reset_index(inplace=True)
top_import_of_poor_exports
```

```python
# calculons les 2 ratios 

df_200['autre_sur_dispo'] = df_200['autre_util'] / df_200['dispo_int']
df_200['nour_sur_nour_anim'] = df_200['alim_anim'] / (df_200['alim_anim'] + df_200['nour'])

df_200.sort_values(by='nour_sur_nour_anim', ascending=False, inplace=True)

display(df_200)
```
